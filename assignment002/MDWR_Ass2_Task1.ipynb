```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# load dataframe from csv file; specify datatypes in columns with mixed types
df = pd.read_csv("dft-road-casualty-statistics-collision-2023.csv", dtype={"accident_index": str, "accident_reference": str})

# display the dataframe
display(df)
```

```python
# display the number of columns
print(len(df.columns))

# display the number of rows
print(len(df))

# find the number of non-null values in each column
df.count()
```

```python
# display the number of distinct values in an attribute
for col in df.columns:
  print(col, ' has (', len(df[col].unique()), ') unique values')
```

```python
# display the data type of each attribute
dataTypeSeries = df.dtypes
for col_idx in range(len(df.columns)):
  print(df.columns[col_idx], 'has type (', dataTypeSeries.iloc[col_idx], ')')
```

```python
# find max and min of numerical attributes
for col_idx in range(len(df.columns)):
  if (not (dataTypeSeries.iloc[col_idx] == 'object')):
    print(df.columns[col_idx], 'has Min =', np.min(df[df.columns[col_idx]]), 'and Max =', np.max(df[df.columns[col_idx]]))
```

```python
# find mean, median and standard deviation of numerical attributes
for col_idx in range(len(df.columns)):
  if (not (dataTypeSeries.iloc[col_idx] == 'object')):
    print(df.columns[col_idx], 'has mean =', round(np.mean(df[df.columns[col_idx]]), 2), 'median =', round(df[df.columns[col_idx]].median(), 2), 'and standard deviation =', round(np.std(df[df.columns[col_idx]]), 2))
```

```python
# create a bar chart for the amount of collisions per month
freq_by_month = df["date"].groupby(df["date"].dt.month).count().plot(kind="bar")
freq_by_month.set_xlabel("Month")
freq_by_month.set_ylabel("Frequency")
```

```python
# create a histogram for the amount of collisions and at which speed limit it happened
hist_speed_limit = df["speed_limit"].hist()
hist_speed_limit.set_xlabel("Speed limit")
hist_speed_limit.set_ylabel("Frequency")
```
